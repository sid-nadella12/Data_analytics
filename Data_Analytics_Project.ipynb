```python
import pandas as pd
import numpy as np
```

```python
crashes_data = pd.read_csv("C:/Users/Dell/Downloads/OneDrive-2021-02-14/crashes.csv", low_memory = False)
vehicles_data = pd.read_csv("C:/Users/Dell/Downloads/OneDrive-2021-02-14/crashes_vehicles.csv", low_memory = False).drop('CRASH_DATE', axis = 1)
```

```python
crashes_data
```

```python
data = pd.merge(crashes_data, vehicles_data, on = 'CRASH_RECORD_ID').reset_index()
```

```python
data
```

```python
print('total number of reported crashes: ', (data['CRASH_RECORD_ID'].nunique()))
```

```python
# changing the date format
data['CRASH_DATE'] = pd.to_datetime(data['CRASH_DATE'])
crashes_data['CRASH_DATE'] = pd.to_datetime(crashes_data['CRASH_DATE'])
```

```python
# Subset DataFrame to include only crashes from 2018, 2019 and 2020, because other years contain only partial data.
data = data[(data['CRASH_DATE'] >= '2018-01-01') & (data['CRASH_DATE'] <= '2020-12-31')]
crashes_data = crashes_data[(crashes_data['CRASH_DATE'] >= '2018-01-01') & (crashes_data['CRASH_DATE'] <= '2020-12-31')]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# sns.set_theme(style = 'darkgrid')

data['Hour'] = data['CRASH_DATE'].dt.hour

# Plot number of crashes by hour of day.
plt.figure(figsize=(15,8))
s = sns.barplot(data=data.groupby('Hour')['CRASH_RECORD_ID'].nunique().reset_index(), x='Hour', y='CRASH_RECORD_ID', palette='GnBu_r', linewidth=0)
s.set_title('Hourly Number of Reported Crashes in Chicago (2018 - 2020)', y=1.02, fontsize=14)
s.set_xlabel('Hour of Day', fontsize=13, labelpad=15)
s.set_ylabel('Number of Crashes', fontsize=13, labelpad=15)
plt.show()
```

```python
# Subset the DataFrame for crashes that occurred during daylight.
df_daylight = data[data['LIGHTING_CONDITION'] == 'DAYLIGHT']

# Determine the make of vehicle involved in the most crashes.
df_daylight['MAKE'].value_counts().nlargest(1)
```

```python
crashes_data.groupby('FIRST_CRASH_TYPE')['WEATHER_CONDITION'].apply(lambda x: x.value_counts().head(1)).reset_index(name='COUNT').rename(columns={'level_1': 'WEATHER'})
```

```python
# Display a plot showing the primary contributing cause of reported crashes.
plt.figure(figsize=(15, 15))
sns.countplot(data=crashes_data,  y='PRIM_CONTRIBUTORY_CAUSE', order = crashes_data['PRIM_CONTRIBUTORY_CAUSE'].value_counts().index)
plt.title('Primary Contributing Cause of Reported Crashes in Chicago (2018 - 2020) ', y=1.01, fontsize=14)
plt.xlabel('Number of Crashes', fontsize=13, labelpad=15)
plt.ylabel('Primary Contributing Cause', fontsize=13, labelpad=15)
plt.show()
```

```python
# Subset to crashes involving a hit and run and drop records without location coordinates.
df_hitrun = crashes_data[crashes_data['HIT_AND_RUN_I'] == 'Y']
df_hitrun = df_hitrun[df_hitrun['LONGITUDE'].notna()]

import folium
from folium import plugins

# Create a map centered on Chicago.
m = folium.Map(location=[41.8781, -87.6798], zoom_start=12)

# Instantiate a mark cluster object for the hit and run crashes.
hitrun = plugins.MarkerCluster().add_to(m)

# Display only crashes where hit and run was reported.
for lat, lng in zip(df_hitrun['LATITUDE'], df_hitrun['LONGITUDE']):
    folium.Marker(
        location=[lat, lng],
           icon=None,
    ).add_to(hitrun)

# Display the map.
m
```
